```python
# Test 1 skapa formerna

from shape import Shape
from circle import Circle
from rectangle import Rectangle

r1 = Rectangle(0, 0, 3, 2)
r2 = Rectangle(2, 3, 5, 5)
c1 = Circle(0, 0, 1)
c2 = Circle(1, 2, 3)

print(r1)
print(r2)
print(c1)
print(c2)
```

```python
# Test 2 area och omkrets

print("r1 area:", r1.area)
print("r1 omkrets:", r1.perimeter())

print("r2 area:", r2.area)
print("r2 omkrets:", r2.perimeter())

print("c1 area:", c1.area)
print("c1 omkrets:", c1.perimeter)

print("c2 area:", c2.area)
print("c2 omkrets:", c2.perimeter)
```

```python
# Test 3 Jämför formerna baserat på area

print("r1 == r2:", r1 == r2)
print("r1 < r2:", r1 < r2)
print("r1 > c1:", r1 > r2)
print("c1 <= c2:", c1 <= c2)
```

```python
# Test 4 

print("Före move:", r1)
r1.move(3, -2)
print("Efter move:", r1)

print("Är r2 en kvadrat?:", r2.is_square())
print("Är c1 en enhetscirkel?:", c1.is_unit_circle())
```
